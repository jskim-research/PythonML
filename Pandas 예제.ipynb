```python
# GroupBy

titanic_groupby = titanic_df.groupby(by="Pclass") # DataFrameGroupBy 객체 반환
print(type(titanic_groupby))
print(titanic_groupby)
```

```python
titanic_df.groupby("Pclass")[["Age", "Name"]].count() # Pclass를 기준으로 그룹했을 때 각 column의 개수
# titanic_df.groupby("Pclass")["Pclass"].count() == titanic_df["Pclass"].value_counts()
```

```python
# Select max(Age), min(Age) from titanic_table group by Pclass

titanic_df.groupby("Pclass")[["Age", "Fare"]].agg([max, min]) # 다수의 통계를 한번에 볼 수 있음
```

```python
# dictionary를 이용하여 다양한 aggregation 함수 적용

agg_format = {"Age":"max", "SibSp":"sum", "Fare":"mean"}
titanic_df.groupby("Pclass").agg(agg_format)
```

```python
# 결손 데이터 (NaN) 처리

# isna() : DataFrame의 컬럽값들이 NaN인지 True/False 값 반환
# fillna() : missing data를 인자로 주어진 값으로 대체

titanic_df.drop(["Age_10"], axis=1).isna().sum() # NaN의 개수
titanic_df.drop(["Age_10"], axis=1).fillna(titanic_df["Age_10"].mean()).isna().sum()
```

```python
# 판다스 람다식 적용하여 데이터 가공
# lambda x : x ** 2 => x를 인자로 받아 x의 제곱 반환 

titanic_df["Name"].apply(lambda x : len(x)).head() # titanic_df["Name_len"] = 
# record 별로 lambda 식 적용
```

```python
titanic_df["Age_cat"] = titanic_df["Age"].apply(lambda x : "Child" if x <= 15 else ("Adult" if x <= 60 else "Elderly"))
titanic_df["Age_cat"].value_counts()
```

```python
# 함수를 통해 람다식을 간략하게 할 수도 있음
def get_category(age):
    cat = ""
    if age <= 5: cat = "Baby"
    elif age <= 12: cat = "Child"
    else: cat = "Adult"
        
    return cat

titanic_df["Age_cat"] = titanic_df["Age"].apply(lambda x : get_category(x))
titanic_df["Age_cat"].value_counts()
```
